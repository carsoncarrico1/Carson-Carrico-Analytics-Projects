```python
path1 = 'C:\\Users\\carso\\Downloads\\qb games.csv'
qb_games = pd.read_csv(path1)
```

```python
X1 = qb_games.drop(qb_games[["Games"]], axis = 1)
Y1 = qb_games["Games"]
```

```python
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size = 0.3, random_state = 6)
```

```python
# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X1_train, Y1_train)

# Make predictions on the test data
y_pred = model.predict(X1_test)

# Calculate and print the model's performance metrics
mse = mean_squared_error(Y1_test, y_pred)
r2 = r2_score(Y1_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared (Coefficient of Determination): {r2}')

for i, predicted_value in enumerate(y_pred):
    print(f"Row {i + 1}: Predicted Value = {predicted_value}")
```
